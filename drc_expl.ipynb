OK, we have its name, but again, what is it?
Well, Python is an object-oriented language, which in simple terms means it has _‘special functions’_ called **Classes**. These Classes allow the creation of _types_ of variables, and **pandas.core.frame.DataFrame** is an example. From now on, we will simply call it **DataFrame** or **pandas DataFrame**, and it is basically analogous to DataFrames in languages like **R** or an **Excel sheet**.

A DataFrame is flexible, so it can store different **types of variables** in each column. We can check the **types** of each column with the command **DataFrame.dtypes** and, in fact, it is the first thing I recommend doing as soon as a file (CSV or Excel) is read.

```python
df.dtypes
```

Another command I constantly use is **DataFrame.describe()** because it calculates the basic statistics of our DataFrame, missing values and can serve as a clue to the distribution of our data.

*   This method (its real name) will only show columns with numeric values.
*   The row with the **count** index can help detect missing data, as the count will be different between columns.

```python
df.describe()
```

Now we will set the **Year** column as the index. This will make it easier to create certain graphs.

```python
df.set_index('Year', inplace=True)
df.head()
```

I have already worked with this csv, so I know the structure. Therefore, I will give you the list of column names that we will be using.

```python
column_names=["Degradation not followed by a deforestation",
              "Degradation followed by a deforestation",
              "Deforestation after degradation",
              "Direct deforestation (excluding conversion to water or plantation)",
              "Forest conversion to plantation",
              "Deforestation to water",
              "Forest regrowth"]
```

We will also choose a country to work with. In this case it will be **Democratic Republic of the Congo**

```python
df_congo = df.loc[df["Country"]=="Democratic Republic of the Congo", column_names]
df_congo.head()
```

####Now we will start ploting. _Pandas_ implements a direct way to work with _matplotlib_, so  we can use it directly on our **DataFrames**. We can also indicate the figure size and give some basic labels to the x and y axis.

Here we will use seaborn for the first time, not as a plotting tool, but as a aesthetic help. I really recomend using the command **sns.set_theme()** because it helps avoiding customizations and saves time.

```python
sns.set_theme()
```

```python
ax=df_congo.plot(kind='bar', stacked=True, figsize=(13,7))
ax.set_ylabel("Disturbance Area (Mha)")
ax.set_xlabel("Year", labelpad=20)
```

You probably noticed that even though we have 7 columns stacked, in most cases only 4 divisions are visible per bar. If we look at our _DataFrame_ **df_congo** we will see that the last 3 columns are at least a hundred times smaller than the others. That's why it could be better to separate this 3 columns in another graph.

```python
df_rest = df_congo.iloc[:,4:7]
df_principal = df_congo.iloc[:,0:4]
df_principal.head()
```

This time we will use a custom palette from seaborn to plot. You'll also notice that, differently to the last plot, the bars touch each other. This is done by setting the argument **width**  to 1. The **x labels** where rotated 45° with the argument _labelrotation_ in **ax.tick_params()**.

```python
sns.set_palette('plasma')
ax=df_principal.plot(kind='bar', stacked=True, width=1, figsize=(13,7))
ax.tick_params(axis='x', labelrotation=45)
ax.set_ylabel("Disturbance Area (Mha)")
ax.set_xlabel("Year", labelpad=20)
```

Lastly, we will plot the 3 variables separated before. The first we got to do is have a look at the variables to make sure they are in the same scale. Looking at the 5 first rows it does seem like the 3 columns are simillar and in the same scale, so we will plot them.

We can use the **DataFrame.describe()** method to have a descriptive view of the data and see where is the problem without plotting. This is not the case for the column **Forest regrowth**, which has a mean at least 20 times higher than the others and a standard deviation almost a 100 times higher than the others.

```python
df_rest.describe()
```

Now we know that the columns _Forest conversion_ and _Deforestation to water_ are in a different scalet than _Forest regrowth_. So, how can we fix this? (and plot this 3 varibles). We could do 2 different plots, one for _Forest conversion_ and _Deforestation to water_, and another for _Forest regrowth_, but we can do something better.

We will plot the variables using a different scale for each group. So, we will have 2 barplots in the same figure, but one will have a scale in the left axes and the other will use the right axes. This will be done by using a special matplotlib method called **Axes.twinx()** that can do exaclty what we want.

To achieve this, first we will divide **df_rest** in  2 different DataFrames and plot them.

```python
df_wt_and_plt = df_rest.iloc[:,0:2]
df_regrowth = df_rest.iloc[:,2]
```

Now we can start plotting. As you can see, we only

```python
##First axes plot
ax=df_wt_and_plt.plot(kind='bar', stacked=True, width=1, figsize=(13,7))
ax.set_ylabel("Disturbance Area (Mha)")
ax.set_xlabel("Year", labelpad=20)
ax.set_xticklabels(ax.get_xticks(), rotation = 45)

##Second axes creation
ax2=ax.twinx()
ax2=df_regrowth.plot(kind='bar', width=1, color="red")
ax2.set_ylabel("Regrowth Area (Mha)", labelpad=20)
```

We will make some style changes:

1.   On the first Axes
    * Multiply **df_wt_and_plt** by 1000, to convert the units from _Mha_ to _Kha_.
    * Use a different color palette by adding the argument **color**.
    * "Manually" modify the legend properties to place it in the left corner.

2.   About the twin Axes
    * This plotbar will only have borders. That's why we use the argument  **edgecolor** as _forestgreen_ and **color** as _None_.
    * Adding the method **Axes.grid(False)** helps to only use the grid of the first **Axes** plot.
    * Also modify the legendo to place it in the right corner.



```python
ax=(df_wt_and_plt*1000).plot(kind='bar', stacked=True, width=1, figsize=(13,7), color=['gold', 'deepskyblue'])
ax.legend(loc='upper left', bbox_to_anchor=(0, 1.07), ncols=3, markerscale=0.1, fontsize=8.3)
ax.set_ylabel("Disturbance Area (Kha)")
ax.set_xlabel("Year", labelpad=20)
ax.set_xticklabels(ax.get_xticks(), rotation = 45)

ax2=ax.twinx()
ax2=df_regrowth.plot(kind='bar', width=1, edgecolor='forestgreen', color='None', lw=3)
ax2.grid(False)
ax2.legend(loc='upper left', bbox_to_anchor=(0.85, 1.07), ncols=3, markerscale=0.1, fontsize=8.3)
ax2.set_ylabel("Regrowth Area (Mha)", rotation=270, labelpad=20)
```
